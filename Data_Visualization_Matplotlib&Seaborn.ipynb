To remove species column from the data set or we want only numerical data no categorical data then we use drop() function

```python
df=iris.drop(['species'], axis = 1) #here axis=1 means  we need to drop entire column and axis=0 means to remove that particular row
```

```python
df.head()
```

for bargraph, histogram, and piechart we mainly require numerical data only, no categorical data is required

```python
df.plot(kind='bar')
```

In the above graph, we can see the overlapping on top of each other hence we are unable to find any insights from the graph.

So, in order to get some insights from the data or to visualize the data properly we need to do the breakdown of graph into smaller parts

```python
df.iloc[0].plot(kind='bar') #here we are performing code on single 0th row
```

```python
df.iloc[50].plot(kind='bar') #here we are performing code on single 50th row
```

```python
df.iloc[50].plot(kind='barh') #in order to turn the graph 90degree we need to use barh in place of bar
```

```python
df.iloc[100].plot(kind='barh')
```

```python
df.iloc[0:5].plot(kind='barh') #to visualize first 5 rows or 10 rows then we do slicing inside iloc() function
```

```python
df.iloc[100:104].plot(kind='barh')
```

ADVANCED VISUALIZATION -> Exploratory Data Analysis (EDA)

1. Scatter plot- It is used to visualize 2 columns relationship, this is part of matplotlib

2. Relational plot- It is used to visualize more than 2 columns relationship, this is part of Seaborn.

Among all the plots, relational plot have all the categories specified inside it.

3. Box plot- Box plot is also a part of cat plot

4. Pair plot- It will pair the columns and then it will give you insights

5. Cat plot- Cat plot refers to categorical plot and it will also has family of graphs inside it

6. Heatmap- we will cover this after the basics of ML because it will need more mathematical computation

```python
data=pd.read_csv("/content/tips.csv")
```

```python
data.head()
```

Scatter Plot

```python
data.plot.scatter(x='total_bill', y='tip')
```

```python
data.plot.scatter(x='total_bill', y='tip', color='g') #colour change to green
```

```python
data.plot.scatter(x='total_bill', y='tip', color='r') #colour change to red
```

Relational Plot

```python
snr.relplot(x='total_bill', y='tip', data=data)
```

```python
snr.relplot(x='total_bill', y='tip', hue='sex',data=data) #hue is a color parameter- any column we put inside hue will become colorful
```

```python
snr.relplot(x='total_bill', y='tip', hue='sex', style='smoker', data=data) #style is a parameter which gives shapes to the categories- any column we put inside style, the categories inside will get a shape
```

```python
snr.relplot(x='total_bill', y='tip', hue='sex', style='smoker', col='day', data=data) #col is a parameter which will give numbers of plots same as number of categories in the column
```

```python
snr.relplot(x='total_bill', y='tip', hue='sex', style='smoker', col='day', size='size', data=data)
```

```python
snr.relplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=data)
```

Q. Visualize the relation between total_bill collected and tip for lunch and dinner time using relational plot

```python
snr.relplot(x='total_bill', y='tip', hue='sex', col='time', data=data)
```

Pair Plot

```python
snr.pairplot(data) #it will pair the columns and then it will give you insights
```

Cat plot- Categorical plot

```python
snr.catplot(x='day', y='total_bill', hue='sex', data=data)
```

```python
snr.catplot(x='day', y='total_bill', hue='sex', kind='bar', data=data)
```

```python
snr.catplot(x='day', y='total_bill', hue='sex', kind ='strip', data=data) # strip graph is same as normal cat graph
```

```python
snr.catplot(x='day', y='total_bill', hue='sex', kind ='violin', data=data)
```

Box Plot

```python
snr.catplot(x='day', y='total_bill', hue='sex', kind ='box', data=data)
```
